```python
from stable_matching.TaskWorker import Task,Worker
import random
n = 2
m = 5
```

```python
workers = []
for i in range(m):
    workers.append(Worker(i))
```

```python
tasks = []
for i in range(n):
    r = 3
    R = [random.sample(workers,random.randint(1,5)) for _ in range(r)]
    tasks.append(Task(ID = i, budget = 1.0, R = R, Q = 1.0))
```

```python
for w in workers:
    values = {}
    for t in tasks:
        values[t] = random.random()
    w.set_preference(values)

for t in tasks:
    costs = {}
    for w in workers:
        costs[w] = random.uniform(0.1,0.8)
    t.set_costs(costs,10)
```

```python
r = {}
c = {}
p = {}
for w in workers:
    p[w.id] = [t.id for t in w.preference_list()]

for t in tasks:
    r[t.id] = [[w.id for w in rr] for rr in t.R]
    c[t.id] = {}
    for w in workers:
        c[t.id][w.id] = t.costs[w]
r,c,p
```

```python
from stable_matching.stableMatching import  generalized_da
for t in tasks:
    t.refresh()
for w in workers:
    w.refresh()
generalized_da(tasks,workers)
```
